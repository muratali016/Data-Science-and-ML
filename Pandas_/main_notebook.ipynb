```python
car_missing=pd.read_csv("car-sales-missing-data.csv")
car_missing
```

```python
car_missing["Odometer"].mean()
```

```python
car_missing["Odometer"]=car_missing["Odometer"].fillna(car_missing["Odometer"].mean())#inplace=true if there is no "x=x"  use inplace
car_missing
```

```python
car_missing=car_missing.dropna()#dropna(inplace=True) if there is no "x=x"  use inplace
car_missing
```

```python
car_missing["Price"] = car_missing["Price"].str.replace('[\$\,]|\.\d*', '').astype(int)
car_missing 
```

```python
#column from Series
seats_column=pd.Series([4,4,4,4])
car_missing["seats"]=seats_column
car_missing
```

```python
car_missing["seats"].fillna(4,inplace=True)
car_missing
```

```python
#column from List
fuel_capacity =[17,21,14,14,13,16]
car_missing["fuel_capacity"]=fuel_capacity
car_missing
```

```python
car_missing["total fuel"]=car_missing["Odometer"]/100*car_missing["fuel_capacity"]
car_missing
```

```python
car_missing.drop("total fuel",axis=1,inplace=True) # I did use the inplace instead of the "X=X" !! 
```
