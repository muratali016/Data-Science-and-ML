```python
#--------------------Two options for subplots--------------------
# Option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2)
ax2.scatter(x,x**2)
ax3.bar(dic.keys(),dic.values())
ax4.hist(x)
```

```python
# Option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(x,x**2)
ax[1,0].bar(dic.keys(),dic.values())
ax[1,1].hist(np.random.random(1000))
```

```python
##--------------------------------Plotting from pandas Dataframes--------------------------------
car_sales=pd.read_csv("car-sales.csv")
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,]|\.\d*', '').astype(int)
car_sales
```

```python
car_sales["sale dates"]=pd.date_range("1/1/2022",periods=len(car_sales))
car_sales
```

```python
car_sales["total revenue"]=car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="sale dates",y="total revenue",kind="scatter");
car_sales.plot(x="sale dates",y="total revenue",kind="hist");
car_sales.plot(x="sale dates",y="total revenue",kind="line");
```

```python
#--------------------different data csv--------------------
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=10);
```

```python
heart_disease.plot.line(figsize=(10,30),subplots =True);
```

```python
## Some fun
over_50=heart_disease[heart_disease["age"]>50]
over_50.head(10)
```

```python
## Pyplot method
over_50.plot(kind="scatter",x="age",y="chol",c="target")
```

```python
## OO method (subplots) with pyplot
fix,ax=plt.subplots(figsize=(10,6))
ax.set_xlim([45,100]) ## make a border for X axis
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax,title="levels")
```

```python
## OO method (subplots) from scratch
fig,ax=plt.subplots(figsize=(10,6))
##plot the data 
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"]
,c=over_50["target"]);
ax.set(title=" LEVELS",xlabel="age",ylabel="chol")

ax.legend(*scatter.legend_elements(),title="target");
ax.axhline(over_50["chol"].mean(),linestyle="dashed");
```

```python
##---------------Creating the whole anatomy of the Matplotlib figure---------------

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
ax0.set(title="levels",xlabel="age",ylabel="chol");
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(y=over_50["chol"].mean(),linestyle="--")
ax0.set_xlim([50,100]) #Change X limits

scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(title="max heart rate",xlabel="age",ylabel="thalach,heart rate");
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(y=over_50["thalach"].mean(),linestyle="dashed");

# Add a title tot he figure
fig.suptitle("Heart disease anlysis",fontsize=18,fontweight="bold");
```
